```python
import pandas as pd
import numpy as np
path = "datas/hour.csv"
df = pd.read_csv(path)
## 删除无用的列
df.drop(columns=["instant","dteday","casual","registered"],inplace=True)
## 检查哪些特征需要做独热编码
# for i in df.columns:
#     a = df[i]
#     print(a.value_counts())

## 独热编码,对需要进行独热编码的列编码
from sklearn.preprocessing import OneHotEncoder
hotCoder = OneHotEncoder(sparse=False,handle_unknown="ignore")
hot = hotCoder.fit_transform(df[["season","mnth","hr","weathersit","weekday"]])
hot = pd.DataFrame(hot)
hot

## 删除掉独热编码的列
df.drop(columns=["season","mnth","hr","weathersit","weekday"],inplace=True)
## 多项式扩展
from sklearn.preprocessing import PolynomialFeatures
polyCoder = PolynomialFeatures(degree=2,interaction_only=False,include_bias=False)
poly = polyCoder.fit_transform(df[["temp","atemp","hum","windspeed"]])
poly = pd.DataFrame(poly,columns=polyCoder.get_feature_names())
poly
```

```python
## 标准化
from sklearn.preprocessing import StandardScaler
ssCoder = StandardScaler()
std = ssCoder.fit_transform(poly)
std = pd.DataFrame(std,columns=polyCoder.get_feature_names())
## 删除掉标准化的列
df.drop(columns=["temp","atemp","hum","windspeed"],inplace=True)
## 合并
df = pd.concat(objs=[hot,std,df],axis=1)
df.head()
```

```python
## 构建X和Y
X = df.iloc[:,0:-1]
Y = df.iloc[:,[-1]]
## 划分训练集和测试集
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)

## 训练LinearRegression模型
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred=model.predict(X_test),y_true=Y_test)

## 训练Ridge
for i in [0.0001,0.1,0.1,0.1]:
    from sklearn.linear_model import Ridge
    model = Ridge(alpha=0.7,max_iter=2000)
    model.fit(X_train,Y_train)
    model.score(X_test,Y_test)
    from sklearn.metrics import mean_squared_error
    mean_squared_error(y_pred=model.predict(X_test),y_true=Y_test)
```
